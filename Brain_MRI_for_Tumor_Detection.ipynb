```python
# import some libraries which will be required
import pandas as pd
import numpy as np
import plotly.graph_objects as go
import xlsxwriter as xlsx
import pickle
import glob
from skimage import io
import math
import openpyxl
```

```python
# create an excel sheet for dataset which does not have tumor
workbook1 = xlsx.Workbook('normal.xlsx')
worksheet1 = workbook1.add_worksheet()

# write headers into the worksheet
worksheet1.write('A1', 'Mean')
worksheet1.write('B1', 'Median')
worksheet1.write('C1', 'Variance')
worksheet1.write('D1', 'Std')
worksheet1.write('E1', 'Max_Intensity')
worksheet1.write('F1', 'Min_Intensity')
worksheet1.write('G1', 'Range')
workbook1.close()
```

```python
# go through the normal dataset directory and make a list of all the pictures

# declare a list to save normal images
normal_list = []

# loop in normal images
for normal in glob.glob(r"C:\Users\Administrator\Desktop\Brain_MRI_Tumor_Detection\brain_tumor_dataset\no\*.*"):
    normal_image = io.imread(normal)

    # append images to list
    normal_list.append(normal_image)
print(len(normal_list))
```

```python
# now again loop in the list and then extract the requried features
for x in range(len(normal_list)):
    
    # read image
     image = normal_list[x]

     
     # convert image inot numpy array
     normal_array = np.asarray(image)
     
     # extract features from each image
    
     Mean = np.mean(normal_array)
     print(Mean)
     Median = np.median(normal_array)
     print(Median)
     Variance = np.var(normal_array)
     print(Variance)
     Std = math.sqrt(Variance)
     print(Std)
     Max_Intensity = np.max(normal_array)
     print(Max_Intensity)
     Min_Intensity = np.min(normal_array)
     print(Min_Intensity)
     Range = np.ptp(normal_array)
     print(Range)

     # list of features
     features_list = [Mean, Median, Variance, Std, Max_Intensity, Min_Intensity, Range]

     # load the saved excel file and then save the extracted features
     outworkbook1 = openpyxl.load_workbook('normal.xlsx')


     # activate worksheet
     outworksheet1 = outworkbook1.active

     # append the feature values into worksheet
     outworksheet1.append(features_list)

     # save the extracted features
     outworkbook1.save(filename = 'normal.xlsx', )
```

```python
# now load and read normal dataset
normal = pd.read_excel('normal.xlsx')
normal.head()
```

```python
# add a column and set it to 0 for normal images
normal['Target'] = 0
normal.head()
```

```python
# save the normal dataset
normal.to_excel('final_normal.xlsx', index = False)
```

## Extract Features from images where Brain Tumor is present

```python
# declare a another excel file
workbook2 = xlsx.Workbook('tumor.xlsx')
worksheet2 = workbook2.add_worksheet()

# write headers to tumor excel sheet
worksheet2.write('A1', 'Mean')
worksheet2.write('B1', 'Median')
worksheet2.write('C1', 'Variance')
worksheet2.write('D1', 'Std')
worksheet2.write('E1', 'Max_Intensity')
worksheet2.write('F1', 'Min_Intensity')
worksheet2.write('G1', 'Range')
workbook2.close()
```

```python
# read from tumor directory and make a list

# declare an empty list
tumor_list = []

for y in glob.glob(r'C:\Users\Administrator\Desktop\Brain_MRI_Tumor_Detection\brain_tumor_dataset\yes\*.*'):

    # read image
    tumor_image = io.imread(y)

    # append each image into tumor list
    tumor_list.append(tumor_image)
print(len(tumor_list))
```

```python
# extract features and add them to newly created list
for image in range(len(tumor_list)):

    # read an image
    tumor_image = tumor_list[image]

    # convert into an array
    tumor_array = np.asarray(tumor_image)

    # extract features 
    Mean = np.mean(tumor_array)
    print(Mean)
    Median = np.median(tumor_array)
    print(Median)
    Variance = np.var(tumor_array)
    print(Variance)
    Std = math.sqrt(Variance)
    print(Std)
    Max_Intensity = np.max(tumor_array)
    print(Max_Intensity)
    Min_Intensity = np.min(tumor_array)
    print(Min_Intensity)
    Range = np.ptp(tumor_array)
    print(Range)

    # make a list of extracted features
    tumor_values = [Mean, Median, Variance, Std, Max_Intensity, Min_Intensity, Range]

    # load newly created excel sheet and then add the features
    outworkbook2 = openpyxl.load_workbook('tumor.xlsx')

    # activate excel sheet
    outsheet2 = outworkbook2.active

    # append tumor list into sheet
    outsheet2.append(tumor_values)

    # save excel file
    outworkbook2.save(filename = 'tumor.xlsx')
    
```
